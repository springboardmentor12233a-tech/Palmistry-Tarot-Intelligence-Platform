```python
import os

hands_dir = '/content/palm_dataset/Hands/Hands'
image_files = [f for f in os.listdir(hands_dir) if f.endswith('.jpg')]

sample_path = os.path.join(hands_dir, image_files[0])

# Load image using MediaPipe's image format
mp_image = mp.Image.create_from_file(sample_path)

# Run detection
detection_result = detector.detect(mp_image)

if detection_result.hand_landmarks:
    print(f"Hand detected in {image_files[0]}!")
    print(f"Number of landmarks: {len(detection_result.hand_landmarks[0])}")
    for landmark in detection_result.hand_landmarks[0]:
        print(landmark.x, landmark.y, landmark.z)
else:
    print("No hand detected in this image.")
```

```python
import mediapipe as mp
from mediapipe.tasks import python
from mediapipe.tasks.python import vision

base_options = python.BaseOptions(model_asset_path='hand_landmarker.task')
options = vision.HandLandmarkerOptions(
    base_options=base_options,
    num_hands=1
)
detector = vision.HandLandmarker.create_from_options(options)
print("Hand Landmarker initialized successfully!")
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Load image for visualization
image = cv2.imread(sample_path)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
h, w, _ = image_rgb.shape

# Draw landmarks manually (new Tasks API doesn't have built-in drawing_utils like the old solutions API)
annotated = image_rgb.copy()
for landmark in detection_result.hand_landmarks[0]:
    x_px = int(landmark.x * w)
    y_px = int(landmark.y * h)
    cv2.circle(annotated, (x_px, y_px), 5, (255, 0, 0), -1)

plt.figure(figsize=(6,6))
plt.imshow(annotated)
plt.axis('off')
plt.title(image_files[0])
plt.show()
```

```python
detected_count = 0
failed_images = []

sample_check = image_files[:30]  # test on first 50 to keep it quick

for fname in sample_check:
    path = os.path.join(hands_dir, fname)
    img = mp.Image.create_from_file(path)
    result = detector.detect(img)
    if result.hand_landmarks:
        detected_count += 1
    else:
        failed_images.append(fname)

print(f"Detected: {detected_count}/{len(sample_check)}")
print(f"Failed on: {failed_images}")
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
for ax, fname in zip(axes, failed_images):
    img = Image.open(os.path.join(hands_dir, fname))
    ax.imshow(img)
    ax.set_title(fname)
    ax.axis('off')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
from tqdm import tqdm

all_landmarks = []
failed_all = []

for fname in tqdm(image_files):
    path = os.path.join(hands_dir, fname)
    img = mp.Image.create_from_file(path)
    result = detector.detect(img)

    if result.hand_landmarks:
        row = {'filename': fname}
        for i, lm in enumerate(result.hand_landmarks[0]):
            row[f'x{i}'] = lm.x
            row[f'y{i}'] = lm.y
            row[f'z{i}'] = lm.z
        all_landmarks.append(row)
    else:
        failed_all.append(fname)

df_landmarks = pd.DataFrame(all_landmarks)
print(f"Total processed: {len(image_files)}")
print(f"Successfully detected: {len(df_landmarks)}")
print(f"Failed: {len(failed_all)}")
df_landmarks.head()
```
